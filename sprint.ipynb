Filtragem Colaborativa


Criar matriz usuário-Item

```python
util_mat = df_combined.pivot_table(index = 'userId', columns = 'title', values = 'rating')
util_mat.head(5)
```

Filtragem Colaborativa Usuário-Usuário - Nisto calculamos a similaridade de todos os usuários com o usuário ativo (o usuário para quem a previsão se destina). Em seguida, classificamos e filtramos os N usuários principais para fazer previsões para o usuário ativo. Isso geralmente é muito eficaz, mas exige muito tempo e recursos. Por exemplo, se Dennis e Davis gostam dos mesmos filmes e sai um novo filme que Davis gosta, então podemos recomendar esse filme a Dennis porque Davis e Dennis parecem gostar dos mesmos filmes.

2) Filtragem Colaborativa Item-Item - É semelhante ao CF Usuário-Usuário, só que agora calculamos a similaridade entre itens para recomendar itens semelhantes. Por exemplo. Ao comprar qualquer produto na Amazon, você encontrará esta linha "Usuários que compraram este item também compraram...", então a Amazon usa amplamente item-item CF. Lembre-se de que não estou dizendo que eles usam apenas item-item CF, eles têm técnicas híbridas para melhor atender usuários até mesmo com interesses únicos.

Item-Item CF é muito mais rápido que User-User CF. e em segundo lugar, os perfis dos usuários mudam rapidamente e todo o modelo do sistema tem que ser recalculado, enquanto as classificações médias dos itens não mudam tão rapidamente, e isso leva a distribuições de classificação mais estáveis ​​no modelo, para que o modelo não precise ser reconstruído como muitas vezes.

```python
user_util_matrix = util_mat.copy()
user_util_matrix = user_util_matrix.apply(lambda row: row.fillna(row.mean()), axis=1)
user_util_matrix.head(5)
```

```python
user_util_matrix.T.corr()
```

```python
user_corr_mat = user_util_matrix.T.corr()
corr_user_1 = user_corr_mat.iloc[0]
```

```python
corr_user_1.sort_values(ascending=False, inplace=True)
```

```python
# Os valores NaN são gerados em corr() porque o std dev é zero, o que é necessário para calcular a similaridade de Pearson.
corr_user_1.dropna(inplace = True)
```

```python
# Despreze o primeiro valor corr, pois é o próprio usuário1
top50_corr_users = corr_user_1[1:51]
```

Então, vamos calcular quais avaliações o usuário 1 daria ao filme com a ajuda do vetor de similaridade. E com base nessa classificação, podemos compará-la com uma classificação limite. Se a classificação for mais alta, ela ficará visível para o usuário ativo em sua lista de recomendações.

```python
df_n_ratings.loc[['Twelve Monkeys (a.k.a. 12 Monkeys) (1995)']]
```

```python
top50_users = top50_corr_users.keys()

count = 0
users = list()
for user in top50_users:
    if df_combined[ (df_combined['userId'] == user) & (df_combined['movieId'] == 32) ]['rating'].sum()  :
        count +=1
        users.append(user)

print(count)
```

```python
# Vamos definir uma função para calcular como o usuário1 avaliará o filme
# Usamos a média ponderada de k usuários semelhantes
def predict_rating():
    sum_similarity = 0
    weighted_ratings = 0
    for user in users:
        weighted_ratings += top50_corr_users.loc[user] * df_combined[
            (df_combined['userId'] == user) & 
            (df_combined['movieId'] == 32)
        ]['rating'].sum()
        sum_similarity += top50_corr_users.loc[user]

    print(weighted_ratings / sum_similarity)
```

```python
def predict_rating():
    sum_similarity = 0
    weighted_ratings = 0
    for user in users:
        weighted_ratings += top50_corr_users.loc[user] * df_combined[
            (df_combined['userId'] == user) & 
            (df_combined['movieId'] == 32)
        ]['rating'].sum()
        sum_similarity += top50_corr_users.loc[user]

    return weighted_ratings / sum_similarity
```

```python
predicted_rating = predict_rating()
print(predicted_rating)
```

Bem, há mudanças muito boas na recomendação deste filme para o primeiro usuário, já que a avaliação é muito boa.

Então, é assim que funciona um CF ingênuo baseado em usuário. As classificações previstas são calculadas de forma semelhante para cada usuário (obviamente para os filmes que ele não foi avaliado) e dependendo da classificação limite, o filme é exibido em sua lista de recomendações ou descartado.

2) Filtragem Colaborativa Baseada em Itens
É bastante semelhante ao algoritmo anterior, mas em vez de encontrar a semelhança do usuário, tentamos encontrar a semelhança do filme.
Assim que tivermos a matriz semelhante do filme, podemos facilmente recomendar filmes semelhantes ao usuário que avaliou qualquer filme do conjunto de dados.

```python
# Vamos encontrar filmes semelhantes ao Jurassic Park
df_n_ratings.loc[['Jurassic Park (1993)']]
```

```python
item_util_matrix = util_mat.copy()
item_util_matrix.head(5)
```

```python
item_util_matrix = item_util_matrix.apply(lambda col : col.fillna(col.mean()), axis=0)
item_util_matrix.head(5)
```

```python
item_util_matrix.isna().sum().sum()
```

Isso significa que cada filme é avaliado por pelo menos 1 usuário.

```python
item_util_matrix.corr()
```

Existem muitos valores NaN e isso ocorre porque quando estamos calculando a correlação de Pearson, se o vetor de classificação tiver todos os valores iguais para, por exemplo, -> [3,0, 3,0, 3,0, 3,0, ....], então o Desvio Padrão é zero e a divisão por zero é indefinida e, portanto, sua correlação com qualquer outro vetor de classificação é NaN.

Como existem muitos filmes avaliados apenas por 1 usuário, toda a média da coluna é preenchida com a avaliação desse usuário e, portanto, a correlação de Pearson fornece valores NaN com qualquer outra coluna.

```python
item_corr_matrix = item_util_matrix.corr()
```

```python
jurassic_park_corr = item_corr_matrix['Jurassic Park (1993)']
jurassic_park_corr = jurassic_park_corr.sort_values(ascending=False)
jurassic_park_corr.dropna(inplace=True)
```

```python
movies_similar_to_jurassic_park = pd.DataFrame(data=jurassic_park_corr.values, columns=['Correlation'], 
                                               index = jurassic_park_corr.index)
movies_similar_to_jurassic_park = movies_similar_to_jurassic_park.join(df_n_ratings['total ratings'])
movies_similar_to_jurassic_park.head(5)
```
